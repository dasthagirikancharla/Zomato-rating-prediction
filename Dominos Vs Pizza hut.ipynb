```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('zomato.csv')
```

```python
df.head()
```

```python
df['approx_cost(for two people)'].unique()
```

```python
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].astype(str)
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].apply(lambda x: x.replace(',','.'))
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].astype(float)
df.info()
```

```python
df['rate'].unique()
```

```python
df = df.loc[df.rate !='NEW']
df = df.loc[df.rate !='-'].reset_index(drop=True)
remove_slash = lambda x: x.replace('/5', '') if type(x) == np.str else x
df.rate = df.rate.apply(remove_slash).str.strip().astype('float')
df['rate'].head()
```

# Dominos Vs Pizza Hut analysis

```python
domi = df[df['name']=="Domino's Pizza"]
Hut = df[df['name']=="Pizza Hut"]
```

```python
domi.shape, Hut.shape
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
sns.countplot(x='online_order', data=domi, )
plt.title('Dominos')
plt.subplot(2,2,2)
sns.countplot(x='online_order', data=Hut)
plt.title('Pizza hut')

# online order for Pizza Hut is more compared to Dominos
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
sns.countplot(x='book_table', data=domi, )
plt.title('Dominos')
plt.subplot(2,2,2)
sns.countplot(x='book_table', data=Hut)
plt.title('Pizza hut')

# for both Dominos and pizza hut table booking is not available
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
sns.countplot(x='rest_type', data=domi, )
plt.title('Dominos')
plt.subplot(2,2,2)
sns.countplot(x='rest_type', data=Hut)
plt.title('Pizza hut')

# Resturent type, Dominos is considered as Quick Bites but pizza hut is conisdered as Casual dininga and Quick bites
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x='votes', y='listed_in(city)', data=domi)

# locations vs votes for Dominos
# MG road, brigade road have very low votes and Banashankari, Koramangala 5th block etc have more votes
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x='votes', y='listed_in(city)', data=Hut)

# locations vs votes for pizza hut
# BTM, Marathahalli have very low votes and Rajainagar, old airport etc have more votes
```

```python
sns.barplot(y='votes', x='listed_in(type)', data=domi)
```

```python
sns.barplot(y='votes', x='listed_in(type)', data=Hut)
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
sns.scatterplot(x='rate', y='votes', data=domi)
plt.title('Dominos')

plt.subplot(2,2,2)
sns.scatterplot(x='rate', y='votes', data=Hut)
plt.title('Pizza hut')
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
sns.scatterplot(x='rate', y='votes', data=domi, hue='listed_in(type)')

plt.subplot(2,2,2)
sns.scatterplot(x='rate', y='votes', data=Hut, hue='listed_in(type)')
```

```python
su1=domi[domi['listed_in(type)']=='Delivery']
su2=domi[domi['listed_in(type)']=='Dine-out']

hu1=Hut[Hut['listed_in(type)']=='Delivery']
hu2=Hut[Hut['listed_in(type)']=='Dine-out']
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
# plt.figure(figsize=(8,8))
plt.title('Dominos')
sns.distplot(su1['rate'],color='r',hist=False, label='Delivery')
sns.distplot(su2['rate'],color='b',hist=False,label='Dine-out')

plt.subplot(2,2,2)
# plt.figure(figsize=(8,8))
plt.title('Hut')
sns.distplot(hu1['rate'],color='r',hist=False,label='Delivery')
sns.distplot(hu2['rate'],color='b',hist=False,label='Dine-out')

# Dine out rating for Dominos has more density then Pizza hut and for delivery Dominos have less density compared to pizza hut
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
sns.distplot(domi['approx_cost(for two people)'])
plt.title('Dominos')
plt.subplot(2,2,2)
sns.distplot(Hut['approx_cost(for two people)'])
plt.title('Pizza hut')
# cost for two people in Dominos is about 400rs/-(high density) and 800res/- (low density)
# but in pizza hut it is about 700rs/-
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
sns.distplot(domi['votes'])
plt.title('Dominos')
plt.subplot(2,2,2)
sns.distplot(Hut['votes'])
plt.title('Pizza hut')
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
sns.lineplot(x='rate', y='approx_cost(for two people)', data=domi, color='r')
plt.title('Dominos')
plt.subplot(2,2,2)
sns.lineplot(x='rate', y='approx_cost(for two people)', data=Hut, color='b')
plt.title('Pizza hut')

# This graph is very intresting, in Pizza hut it is almost people paid about 700rs/- are both happy and unhappy
# but in Dominos we can see range of customer who rated from 3 to 4. 
```
